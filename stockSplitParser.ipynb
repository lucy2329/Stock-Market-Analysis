```python
#!/usr/bin/env python3
# coding: utf-8


import requests

headers = requests.utils.default_headers()
headers.update({ 'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:52.0) Gecko/20100101 Firefox/52.0'})


from bs4 import BeautifulSoup

for year in range(2016,2020):

    url = "https://www.moneycontrol.com/stocks/marketinfo/splits/homebody.php?sel_year="+str(year)
    req = requests.get(url,headers)
    soup = BeautifulSoup(req.content, 'html.parser')

    values = []
    dict = {}

    for x in soup.find_all(class_="tim"):
        values.append(x.get_text())

    values = values[1:]
    del values[0::5]
    i=0

    while(i<len(values)):
        dict[values[i]] = [values[i+1], values[i+2], values[i+3]]
        i+=4

    print("Init Dictionary : ",dict)

    links = []
    text = []

    for a in soup.find_all("a", class_="tim"):
        links.append('moneycontrol.com'+a['href'])
        text.append(a.get_text())
    
    print('links :', links)
    print('text  :', text)

    NSE = []
    i=0
    print('len :', len(text))
    while(i<len(text)):
        url = 'http://'+links[i]
        req = requests.get(url,headers)
        if(req.status_code >=400):
            del text[i]
            i+=1
        else:        
            newsoup = BeautifulSoup(req.content, 'html.parser')    

            p = newsoup.find(class_="bsns_pcst")
            span = p.find_all("span")

            counter=0
            for y in span:
                if(counter==1):
                    NSE.append(y.get_text())
                counter+=1
            i+=1
        
        print(i)
        
    
    print("Share names :",NSE)
    print('nse:', len(NSE))
    print('text:', len(text))
    for i in range(0,len(text)):
        dict[text[i]].append(NSE[i])
        
    print("Final Dictionary : ", dict)

    with open('D:\\StockStuff\\'+str(year)+'splits.txt', 'w+') as f:
        for key, value in dict.items():
            if(len(value)==3):
                f.write('%s,%s,%s,%s,%s\n' % (key, value[0],value[1],value[2],'NA'))
            else:
                if(value[3] == ''):
                    f.write('%s,%s,%s,%s,%s\n' % (key, value[0],value[1],value[2],'NA'))
                else:
                    f.write('%s,%s,%s,%s,%s\n' % (key, value[0],value[1],value[2],value[3]))

    f.close()

```
